```python
x = [20,30,34,54,23,45]
y = [400,450,475,342,453,565]
plt.xlabel("current")
plt.ylabel("voltage")

plt.xlim([20,55])
plt.ylim([300,600])
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(5,100,50)
y = []
for i in x:
    y.append(math.log10(i))

plt.plot(x,y,"b+")
plt.title("watt")
plt.xlabel("current")
plt.ylabel("voltage")
```

```python
x = np.linspace(5,100,50)
y = []
for i in x:
    y.append(math.log10(i))
i= np.linspace(5,100,50)
j=i**2
plt.plot(x,y,"g^",i,j,"r+")
plt.show()
```

```python
x= np.linspace(1,10,100)
y= np.log(x)
plt.figure(1)

plt.subplot(131)
plt.title('y = log(x)')
plt.plot(x,y)

plt.subplot(132)
plt.title("y = log(x)**2")
plt.plot(x,y**2)

plt.subplot(133)
plt.title("y = log(x)**3")
plt.plot(x,y**3)

plt.figure(2)
plt.subplot(111)
plt.plot(x,x**2)

plt.show()
```

```python
plt.figure(1)

plt.subplot(2,2,1)
plt.title("y= log(x)")
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title("y= log2(x)")
plt.plot(x,np.log2(x))

plt.figure(2)
plt.subplot(2,2,3)
plt.title("y= log5(x)")
plt.plot(x,(np.log(x))**3)

plt.subplot(2,2,4)
plt.title("y= log(x)**2")
plt.plot(x,(np.log(x))**2)

plt.show()
```
